```python
import tensorflow as tf
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Cargar modelo
model = tf.keras.models.load_model("checkpoints/butterfly_model.keras")

model
```

```python
# Ruta de la imagen
image_path = "monarch.jpg"

# Abrir imagen
img = Image.open(image_path)

# Mostrarla
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Redimensionar al tamaño usado durante entrenamiento
img = img.resize((75, 75))  # cambia si usaste otro tamaño

# Convertir a array
img_array = np.array(img)

# Asegurarse de que sea RGB
img_array = img_array[:, :, :3]

# Normalizar si entrenaste con valores entre 0 y 1
img_array = img_array / 255.0

# Añadir dimensión batch
img_array = np.expand_dims(img_array, axis=0)
```

```python
# Predicción
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions)
predicted_species = class_indices[predicted_class_index]
```
